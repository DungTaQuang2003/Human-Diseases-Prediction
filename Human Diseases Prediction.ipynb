#**Xây dựng mô hình**

##**Train Test Split**

```python
X_train_valid = train_data.drop(columns=['prognosis'])
y_train_valid = train_data['prognosis']
```

```python
X_test = test_data.drop(columns=['prognosis'])
y_test = test_data['prognosis']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_valid, y_train, y_valid = train_test_split(X_train_valid, y_train_valid,
                                                      test_size = 0.2,
                                                      stratify = y_train_valid,
                                                      random_state = 42,
                                                      shuffle=True)
```

##**Build Model**

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
```

```python
tree = DecisionTreeClassifier()
tree.fit(X_train, y_train)
tree.score(X_valid, y_valid)
```

```python
lr = LogisticRegression()
lr.fit(X_train, y_train)
lr.score(X_valid, y_valid)
```

```python
rfc = RandomForestClassifier(criterion='entropy')
rfc.fit(X_train, y_train)
rfc.score(X_valid, y_valid)
```

**Nhận xét:**
- Hai mô hình Logistic Regression và Random Forest đều cho kết quả độ chính xác khá là cao, hoàn hảo là 1.0, hay 100%. Tuy nhiên, đây cũng không hẳn là một tín hiệu đáng mừng mà nó chính là một dấu hiệu có thể gây nghi ngờ về việc mô hình có thể đang bị học quá khớp (Overfitting) trên tập huấn luyện hoặc tập xác thực.
- Đối với mô hình Decisin Tree thì kết quả độ chính xác cho ra thấp hơn, xấp xỉ 0.7. Kết quả này tuy là thấp hơn cả so với 2 mô hình trên, nhưng cũng chưa thể nào mà kết luận ngay rằng là khả năng phân loại của nó kém hơn so với 2 mô hình còn lại.

=> Qua nhận xét ban đầu về khả năng phân loại của cả 3 mô hình thì ta thấy vẫn chưa thể kết luận được mô hình nào là tốt nhất trong tất cả, do vậy trong bước tiếp theo ta sẽ cần phải kiểm chứng, tinh chỉnh lại mô hình và kiểm tra thêm tất cả 3 mô hình để xem xem là mô hình nào hiệu quả và tối ưu nhất với bài toán đặt ra rồi từ đó mới có thể đưa ra kết luận đúng đắn là mô hình nào sẽ phù hợp.

##**Điều chỉnh mô hình**

```python
from sklearn.model_selection import GridSearchCV, StratifiedKFold
```

```python
from sklearn.metrics import classification_report
```

```python
skf = StratifiedKFold()
```

**Decision Tree**

```python
modelDt = DecisionTreeClassifier(criterion='entropy')
params = {
    'max_depth': range(5,50),
    'max_leaf_nodes': range(5,25),
    'min_samples_leaf': range(2,5)
}
grid_search = GridSearchCV(estimator = modelDt, param_grid = params, cv = skf, scoring = 'f1_macro', n_jobs = -1)
```

```python
grid_result = grid_search.fit(X_train, y_train)
```

```python
print(grid_result.best_params_)
print(grid_result.best_score_)
```

```python
dt_model = modelDt.set_params(**grid_result.best_params_)
dt_model.fit(X_train, y_train)
```

```python
dt_model.score(X_valid, y_valid)
```

```python
y_valid_pred = dt_model.predict(X_valid)
print(classification_report(y_valid, y_valid_pred))
```

```python
dt_model.score(X_test, y_test)
```

```python
y_pred = dt_model.predict(X_test)
print(classification_report(y_test, y_pred))
```

**Nhận xét:** Sau khi được cắt tỉa, điều chỉnh lại các tham số thì ta thấy kết quả của mô hình Decision Tree còn giảm xuống đáng kể hơn so với kết quả của mô hình ban đầu. Điều này cho thấy là mô hình sau khi tinh chỉnh này đang bị Underfitting, không có khả năng học tốt từ tập dữ liệu, kết quả đánh giá f1-score = 0.5 và accuracy = 0.57 cho thấy khả năng phân loại của mô hình không tốt nên không thể ứng dụng cho bài toán được.

**Random Forest**

```python
modelRf = RandomForestClassifier(criterion='entropy')
params = {
    'n_estimators': range(10, 20),
    'max_depth': range(5,10),
    'max_leaf_nodes': range(5,25),
    'min_samples_leaf': range(2,3)
}
grid_search = GridSearchCV(estimator = modelRf, param_grid = params, cv = skf, scoring = 'f1_macro', n_jobs = -1)
```

```python
grid_result = grid_search.fit(X_train, y_train)
```

```python
print(grid_result.best_params_)
print(grid_result.best_score_)
```

```python
rf_model = modelRf.set_params(**grid_result.best_params_)
rf_model.fit(X_train, y_train)
```

```python
rf_model.score(X_valid, y_valid)
```

```python
y_valid_pred = rf_model.predict(X_valid)
print(classification_report(y_valid, y_valid_pred))
```

```python
rf_model.score(X_test, y_test)
```

```python
y_pred = rf_model.predict(X_test)
print(classification_report(y_test, y_pred))
```

**Nhận xét:**
Đối với mô hình Random Forest, sau khi đã tinh chỉnh, cắt tỉa lại cho tất cả các cây trong rừng và giới hạn độ sâu của ta thấy thì mô hình đã tránh được hiện tượng Overfitting so với mô hình ban đầu và thu được kết quả phân loại cũng khá là tốt và ổn định:
- Với tập Validation: accuracy = 0.92 và f1-macro = 0.85
- Với tập Test: accurary = 0.95 và f1-macro = 0.94.

**LogisticRegession**

- Trong mô hình này, ta sẽ không sử dụng penalty là None vì nó chỉ dùng trong trường hợp:
    - Dữ liệu lớn và đa dạng(dữ liệu của ta có ở mức khá nhỏ mặc dù khá đa dạng)
    - Không có dấu hiệu của overfitting(ở đây ta đang có dấu hiệu overfitting)
    - Mục tiêu chính của mô hình có thể là khớp tốt với dữ liệu huấn luyện mà không cần phải lo lắng về việc tổng quát hóa tốt cho dữ liệu mới(ở đây ta cần đoán bệnh)
- Khi không dùng 2 penalty trên ta chỉ còn 'l1' và dùng với solver là: liblinear và saga. Nhưng liblinear chỉ dùng cho dữ liệu có nhãn là nhị phân nên đây không phải là lựa chọn phù hợp.<br>

**-> Do đó trong mô hình tinh chỉnh, ta sẽ sử dụng các tham số còn lại là: penalty='l1' hoặc 'elasticnet', solver='saga' và multi_class='multinomial'**

```python
model_lr = LogisticRegression(multi_class='multinomial') # vì bài toán đặt ra là phân loại đa lớp nên ta sẽ sử dụng multi_class='multinomial'

params = {
    'penalty': ['l1', 'elasticnet'], # Do ban đầu có thử với l2 và nó bị overfitting nên em đã loại bỏ đi
    'solver': ['saga'],
}

grid_search = GridSearchCV(estimator=model_lr,
                           param_grid=params,
                           cv=skf,
                           scoring='f1_macro',
                           n_jobs=-1)
```

```python
grid_result = grid_search.fit(X_train, y_train)
```

```python
print(grid_result.best_params_)
print(grid_result.best_score_)
```

```python
lr_model = model_lr.set_params(**grid_result.best_params_)
```

```python
lr_model.fit(X_train, y_train)
```

```python
lr_model.score(X_valid, y_valid)
```

```python
y_valid_pred = lr_model.predict(X_valid)
print(classification_report(y_valid, y_valid_pred))
```

```python
lr_model.score(X_test, y_test)
```

```python
y_pred = lr_model.predict(X_test)
print(classification_report(y_test, y_pred))
```

**Nhận xét:**
Với mô hình Logistic Regression, sau khi đã tinh chỉnh lại các tham số sao cho phù hợp thì mô hình cũng đã tránh được hiện tượng Overfitting so với mô hình ban đầu và thu được kết quả phân loại cũng rất là tốt và ổn định, tốt nhất trong 3 mô hình thử nghiệm:
- Với tập Validation: accuracy = 0.92 và f1-macro = 0.86
- Với tập Test: accurary = 0.98 và f1-macro = 0.98

Cuối cùng, dựa trên các tiêu chí để lựa chọn một mô hình phù hợp với bài toán:
+ Hiệu suất thống kê
+ Dễ giải thích
+ Hiệu suất vận hành

thì ta thấy rằng mô hình Logistic Regression đang đáp ứng được nhiều hơn cả nên mô hình cuối cùng ta chọn cho bài toán này là Logistic Regression.
